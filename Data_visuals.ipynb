```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("bar_chart_data.csv")
data
```

```python
plt.figure(figsize=(9,6))
plt.bar(data["Brand"],data["Cars Listings"],align='center',color='grey')
plt.xticks(rotation=45, fontsize=13)
plt.ylabel("Number of sold car")
plt.title("Car Brands with Sell" , fontsize=16, fontweight="bold")
plt.show()
```
